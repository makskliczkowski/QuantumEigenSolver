```python
from QES.general_python.algebra.utils import get_backend as __backend
from QES.general_python.algebra.ran_wrapper import uniform, randint

# Test random
seed        = None
backend     = 'np'
for i in range(5):
    print(f"Iteration {i + 1}:")
    print("-" * 40)
    modules = __backend(backend, random=True, seed=seed)
    backend, (rnd_module, key) = modules if isinstance(modules, tuple) else (modules, (None, None))
    print(f"Backend: {backend}")
    print(f"Key: {key}")
    print(f"Random Module: {rnd_module}")
    print(f"Random Value: {rnd_module.random()}")
    print(f"Random Integer [0, 10): {rnd_module.integers(0, 10)}")
    print(f"Random Uniform [0, 10): {rnd_module.uniform(0, 10)}")
    print(f"Random Normal (mean=0, std=1): {rnd_module.normal(0, 1)}")
    print(f"Uniform Function Output [0, 10): {uniform(shape=(5,), minval=0, maxval=10, backend=backend, seed=seed)}")
    print("-" * 40)
    print()
    
backend = 'jax'
for i in range(5):
    print(f"Iteration {i + 1}:")
    print("-" * 40)
    modules = __backend(backend, random=True, seed=seed)
    backend, (rnd_module, key) = modules if isinstance(modules, tuple) else (modules, (None, None))
    print(f"Backend: {backend}")
    print(f"Key: {key}")
    print(f"Uniform Function Output [0, 10): {uniform(shape=(5,), minval=0, maxval=10, backend=backend, seed=seed)}")
    print("-" * 40)
    print()
```
